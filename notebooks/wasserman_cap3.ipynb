```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
```

First, we need to compute the values of mean and variance for a uniform random variable in (0,1):$$\mathbb{E}X = \int_a^bx\frac{1}{b-a}dx = \frac{a+b}{2}$$ e $$\mathbb{V}ar(x) = \mathbb{E}X^2-(\mathbb{E}X)^2 = \frac{(a-b)^2}{12}$$

```python
a, b = 0, 1 #lower and upper boundaries of interval
true_mean = (a+b)/2
true_var = (a-b)**2/12
true_mean, true_var
```

```python
lista = [100, 500, 2500, 10000]
lista_mean = []
lista_var = []
for i in lista:
  sample = np.random.uniform(a,b,i) #generate a sample of size i uniformly distributed in (a,b)
  sample_mean = np.mean(sample)#compute the sample mean
  lista_mean.append(sample_mean)
  sample_var = np.var(sample)#compute the sample variance
  lista_var.append(sample_var)

plt.plot(lista, lista_mean, label='mean')
plt.plot(lista, lista_var, label='variance')
plt.plot(lista, true_mean*np.ones(len(lista)), label='true_mean')
plt.plot(lista, true_var*np.ones(len(lista)), label='true_var')
plt.legend()
plt.show()
```

```python
n = 5000
sample = np.random.uniform(a,b,n)
sample_mean = np.mean(sample)
_, bins, _ = plt.hist(sample, density=True)
plt.plot(bins, np.ones(bins.shape),linewidth=2, color='r')
plt.show()
```

```python
def clt(sample_data,sample_size,total_samples):
  #given my sample of 5000 uniform dist. i want to generate 100 samples of sample means of shape 30
  sample_mean = []
  for a in range(total_samples):
    sample = np.random.choice(sample_data, size = sample_size)    #generates a sample of size sample_size
    mean = np.mean(sample)      #computing sample mean
    sample_mean.append(mean)

  return sample_mean
```

```python
sbn.displot(clt(sample, 3000, 10000), kind="kde")
```
